```python
import pandas as pd
import numpy as np
```

Changing Options with Attributes and Dot Syntax

```python
data=np.random.randint(0,100,[1000,50]) #1000 rows and 50 columns and all values will be between 0 to 100 n np is library random is module and ranint is method
df=pd.DataFrame(data) #dataframe constructor method which is avaiable on pandas library,it is going to turn an numpy array into dataframe
df.tail(3)
```

```python
pd.options.display.max_rows=4
```

```python
df
```

```python
pd.options.display.max_columns
```

Changing pandas Options with Methods

```python
data=np.random.randint(0,100,[1000,50]) 
df=pd.DataFrame(data)
df.tail(2)
```

```python
pd.get_option("max_rows")
```

```python
pd.get_option("max_columns")
```

```python
pd.set_option("max_rows",16)
```

```python
df
```

```python
pd.reset_option("max_rows")
```

```python
pd.get_option("max_columns")
```

```python
pd.describe_option("max_columns")
```
