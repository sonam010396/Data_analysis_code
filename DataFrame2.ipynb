The .isin() Method #when u have to filter multiple value without creating multiple series

```python
import pandas as pd
df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time"])
df["Senior Management"]=df["Senior Management"].astype("bool")
df["Gender"]=df["Gender"].astype("category")
df.head(3)
```

```python
mask1= df["Team"]=="Legal"
mask2=df["Team"]=="Sales"
mask3=df["Team"]=="Product"
df[mask1 | mask2 & mask3]
```

```python
mask=df["Team"].isin(["Legal","Sales","Product"])
df[mask]
```

The .duplicate() Method

```python
df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time"])
df["Senior Management"]=df["Senior Management"].astype("bool")
df["Gender"]=df["Gender"].astype("category")
df.sort_values("First Name",inplace = True)
df.head(3)
```

```python
df["First Name"].duplicated(keep="first")# it will keep the first duplicate value and other  same value will be call as duplicate
```

```python
df["First Name"].duplicated(keep=False)# it will mark all the duplicate value as True
```

```python
mask=~df["First Name"].duplicated(keep=False)# tild symbol makes true as false and false as true
```
