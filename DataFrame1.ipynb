Drop Rows with Null Values

```python
nba=pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.dropna()
```

```python
nba.dropna(how="all")# it removes the row which has all null values
```

```python
nba.dropna(axis=1)#it will remove the columns which contains the null value
```

```python
nba.dropna(subset=["Salary"])# it will only drop the row in salary column that has null value
```

The .astype() Method

```python
nba=pd.read_csv("nba.csv").dropna(how="all")
nba.tail(3)
nba["Salary"].fillna(0,inplace =True)
nba["College"].fillna("NO COLLEGE",inplace =True)
```

```python
nba["Salary"]=nba["Salary"].astype("int")
```

```python
nba["Position"].nunique()   
nba["Position"].astype("category")#Category is a datatype in Pandas (we use when u have small number of unique value)
```

Rank values with the .rank() Method

```python
nba=pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"]=nba["Salary"].fillna(0).astype("int")
nba.head(3)
```

```python
nba["Salary"]=nba["Salary"].rank(ascending=False).astype("int")
```
